```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
path = "./data/company_sales_data.csv"

data = pd.read_csv(path)
data
```

# Q1)

```python
data.set_index('month_number', inplace=True)

plt.plot(data['total_profit'])

plt.yticks(range(100000, 600000, 100000))
plt.xticks(range(1, 13, 1))

plt.xlabel('Month Number')
plt.ylabel('Profit in dollars')

plt.title('Company profit per month')

plt.show()
```

# Q2)

```python
plt.plot(data['total_profit'], ls="dotted", color='red', marker='o', markerfacecolor='red', lw=3)

plt.yticks(range(100000, 600000, 100000))
plt.xticks(range(1, 13, 1))

plt.xlabel('Month Number')
plt.ylabel('Sold units number')

plt.title('Company Sales data of last year')
plt.legend(['Profit data of last year'], loc='lower right')

plt.show()
```

# Q4)

```python
x = range(1, 6)
y = [i**2 for i in x]

plt.plot(x, label='values', marker='o')
plt.plot(y, label='square of values', marker='o')

plt.legend()
plt.show()
```

# Q5)

```python
X = np.linspace(0, 10, 1000)

plt.plot(np.sin(X), label='sin(x)')
plt.plot(np.cos(X), label='cos(x)')

plt.xlabel("Value of X")
plt.ylabel("Value of sin(X) and cos(X)")

plt.legend()

plt.show()
```
