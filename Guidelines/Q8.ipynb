```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.random.normal(170, 10, 250)

plt.hist(x)
plt.show()
```

```python
x = np.arange(0, 12, 0.1)

plt.plot(x, np.sin(x))
plt.plot(x, np.cos(x))

plt.legend(['sin(x)', 'cos(x)'])

plt.xlabel("Value of sin(x) & cos(x)")
plt.ylabel("Value of x")

plt.show()
```
