```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.array([0, 1, 2, 3])
y = np.array([3, 8, 1, 10])

plt.subplot(2, 2, 1)
plt.plot(x, y, color='red')

x = np.array([0, 1, 2, 3])
y = np.array([10, 20, 30, 40])

plt.subplot(2, 2, 2)
plt.plot(x, y, color='black')

x = np.array([0, 1, 2, 3])
y = np.array([3, 10, 13, 20])

plt.subplot(2, 2, 3)
plt.plot(x, y, color='blue')

x = np.array([0, 1, 2, 3])
y = np.array([10, 12, 20, 18])

plt.subplot(2, 2, 4)
plt.plot(x, y, color='yellow')

plt.show()
```
